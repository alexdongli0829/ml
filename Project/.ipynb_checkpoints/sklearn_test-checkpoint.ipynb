```python
import os
import pandas as pd
```

```python
train_path='/home/ubuntu/learn-with-other-kaggle/data/train.csv'
test_path='/home/ubuntu/learn-with-other-kaggle/data/test.csv'

train_data=pd.read_csv(train_path, index_col = 'Id')
test_data=pd.read_csv(test_path, index_col = 'Id')


train_data.head()
```

```python
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import mean_absolute_error, confusion_matrix
from xgboost import XGBClassifier
from sklearn.feature_selection import SelectFromModel
```

```python
#get the unique value number for each column
train_data.nunique()
```

```python
# separate the features with the labels
features_train = train_data.drop('Cover_Type', axis=1)
label_train = train_data['Cover_Type']
features_test = test_data
```

```python
x_train, x_valid, y_train, y_valid = train_test_split(features_train, label_train, test_size=0.1)
```

```python
# random forest classifier

rfc = RandomForestClassifier(n_estimators = 100, random_state=0)
rfc.fit(x_train,y_train)
pred=rfc.predict(x_valid)
mae = mean_absolute_error(y_valid, pred)
```

```python
pred_search=rf_search.predict(x_valid)
mae = mean_absolute_error(y_valid, pred_search)
```

```python
# Referening some blogs and first change the model to StackingCVClassifier to check if this can improve the performance
#before doing any feature engineering
#https://www.kaggle.com/edumunozsala/feature-eng-and-a-simple-stacked-model
#https://www.kaggle.com/stevegreenau/stacking-multiple-classifiers-clustering
#http://rasbt.github.io/mlxtend/user_guide/classifier/StackingCVClassifier/
from mlxtend.classifier import StackingCVClassifier
from lightgbm import LGBMClassifier
```

```python
rf_clf = RandomForestClassifier(n_estimators = 100, random_state=0)
lgbm_clf = LGBMClassifier(n_estimators = 100)
ada_cfl = AdaBoostClassifier(n_estimators = 100)
labels =['rf','lgbm','ada']
models = {'rf':rf_clf, 'lgbm':lgbm_clf, 'ada':ada_cfl}
```

```python
# according to the blog, we need first find out which one is the best classifier using the k-folder and use it as the 
#second layer classfier
from sklearn.model_selection import KFold, cross_val_score

cv = KFold(n_splits=5, shuffle=True, random_state=0)
results = {}
for label, model in models.items():
    print("evaluating model %s"%label)
    #https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.cross_val_score.html
    val_results=cross_val_score(model,X=x_train,y=y_train, cv=cv, scoring='accuracy')
    results[label]=val_results
    print("model %s, val result mean %f, std %f"%(label, results[label].mean(), results[label].std()))

```

```python
model_score={}
# remember to use results.items() intead of results, if use results directly, the label, result will be "r","f"
for label, result in results.items():
    model_score[label] = result.mean()
best_model = max(model_score)
print("The best model is %s, accuracy is %f"%(best_model, model_score[best_model]))
```

```python
print(list(models.values()))
```

```python
# stacking the 3 classifier and in the same time, use the best one for the second layer classifier

# If I use classifiers=models.values(), there will be error: TypeError: can't pickle dict_values objects
# need use list(models.values()) to convert the dict_values to list
stacked_model = StackingCVClassifier(classifiers=list(models.values()),
                             meta_classifier=rf_clf,
                             cv=5,
                             use_probas=True,
                             drop_last_proba=False,
                             use_features_in_secondary=True,
                             verbose=1,
                             store_train_meta_features=True,
                             random_state=0,
                             n_jobs=-1)
```

```python
score = cross_val_score(stacked_model,X=x_train,y=y_train, cv=cv, scoring='accuracy',verbose=0)
```

```python
print("Final score mean %f, std %f"%(score.mean(), score.std()))
```

```python
stacked_model.fit(x_train,y_train)
```

```python
stackClf_result = stacked_model.predict(test_data)
```

```python
# to improve the model, I would use the GridCVSearch to seach the best for the n_estimators
#http://rasbt.github.io/mlxtend/user_guide/classifier/StackingCVClassifier/#example-3-stacked-cv-classification-and-gridsearch

stacked_model_grid_search = StackingCVClassifier(classifiers=list(models.values()),
                             meta_classifier=rf_clf,
                             cv=5,
                             use_probas=True,
                             drop_last_proba=False,
                             use_features_in_secondary=True,
                             verbose=1,
                             store_train_meta_features=True,
                             n_jobs=-1)

# list all the parameters which can be search
stacked_model_grid_search.get_params().keys()

# define the list
paremeters_check={
    "meta_classifier__n_estimators":[100, 200, 300, 400]
    #"lgbmclassifier__n_estimators":[50, 100, 200, 300, 400],
    #"randomforestclassifier__n_estimators":[50, 100, 200, 300, 400],
    #"adaboostclassifier__n_estimators":[50, 100, 200, 300, 400]
}

#https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.GridSearchCV.html
search_model = GridSearchCV(estimator=stacked_model_grid_search, 
                    param_grid=paremeters_check, 
                    cv=5,
                    verbose=1,
                    scoring='accuracy',
                    refit=True)

search_model.fit(x_train, y_train)
```

```python
cv_keys = ('mean_test_score', 'std_test_score', 'params')

#for item in enumerate(["a", "b", "c"]):
#    print item

#(0, "a")
#(1, "b")
#(2, "c")
for r, _ in enumerate(search_model.cv_results_['mean_test_score']):
    print("%0.3f +/- %0.2f %r"
          % (search_model.cv_results_[cv_keys[0]][r],
             search_model.cv_results_[cv_keys[1]][r] / 2.0,
             search_model.cv_results_[cv_keys[2]][r]))

print('Best parameters: %s' % search_model.best_params_)
print('Accuracy: %.2f' % search_model.best_score_)
```

```python
# check the parameters for other classifier
for label, model in models.items():
    print("searching the parameter of %s"%label)
    parameters = {'n_estimators':[100, 200, 300, 400]}
    search_model_l1 = GridSearchCV(estimator=model, 
                    param_grid=parameters, 
                    cv=5,
                    verbose=1,
                    scoring='accuracy',
                    refit=True)
    search_model_l1.fit(x_train,y_train)
    print('Best parameters: %s' % search_model_l1.best_params_)
    print('Accuracy: %.2f' % search_model_l1.best_score_)
```

```python
# After the grid search, I set the n_estimators based on the search result and stacking again
rf_clf = RandomForestClassifier(n_estimators = 200)
lgbm_clf = LGBMClassifier(n_estimators = 400)
ada_cfl = AdaBoostClassifier(n_estimators = 100)
rf_meta = RandomForestClassifier(n_estimators = 100)

labels =['rf','lgbm','ada']
models = {'rf':rf_clf, 'lgbm':lgbm_clf, 'ada':ada_cfl}



stacked_model_with_all_best = StackingCVClassifier(classifiers=[rf_clf,lgbm_clf,ada_cfl],
                             meta_classifier=rf_meta,
                             cv=5,
                             use_probas=True,
                             drop_last_proba=False,
                             use_features_in_secondary=True,
                             verbose=1,
                             store_train_meta_features=True,
                             n_jobs=-1)
stacked_model_with_all_best.fit(x_train,y_train)
```

```python
# check the score for the new model
score = cross_val_score(stacked_model_with_all_best,X=x_valid,y=y_valid, cv=cv, scoring='accuracy',verbose=0)
```

```python
# based on the valid data, it looks the accuracy is a little bitter higher
print("score %f, std %f"%(score.mean(),score.std()))
```

```python
# train the model with all the data
stacked_model_with_all_best.fit(features_train, label_train)
```

```python
stack_with_optimized_estimators_number = stacked_model_with_all_best.predict(test_data)
```
